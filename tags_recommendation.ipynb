```python
posts['Tags'] = posts['Tags'].str.replace(r"\<", "")
posts['Tags'] = posts['Tags'].str.replace(r"\>", ",")
posts['Tags'] = posts['Tags'].map(lambda x: str(x)[:-1])
```

```python
posts['count'] = posts['Tags'].apply(lambda x: len(x.split(',')))
```

```python
def getdata(ques, tagName, posts):
    result = []
    counter_list = []
#     posts['Recommended_tags'] = ""
#     print type(tagName)
    for question in ques:
        count = 0
        counter = 0
        question = str(question)
        question = question.lower()
        word_tokens = question.split() 
        tags_found = []
#         tags_found.append(question)
        for word in word_tokens:            
            if word in tagName.values:
#                 result[count].append(word)
                tags_found.append(word)
                counter = counter+1
        counter_list.append(counter)        
#         posts['Recommended_tags'][count] = counter        
#         posts['Recommended_tags'].iloc(count) = counter
#         print counter
#                 tags_found = sorted(tags_found)
#         print tags_found
        result.append(tags_found)
#                 print word + 'yo'
#             else:
#                 print word + 'bo'
#             if(word in tags['TagName'].values):
#                 print word + 'yo'
#             else:
#                 print word + 'bo'    
#         print ques + result[count]
        count = count + 1   
    
    exact_match_list = []
    for recommend_tags_list in result:
        exact_match = 0
        for recommend_tags in recommend_tags_list:
            if recommend_tags in posts['Tags'].values:
                exact_match = exact_match + 1
        exact_match_list.append(exact_match)      
        print str(recommend_tags_list) + ' ' + str(exact_match)    
    return counter_list,result, exact_match_list
```

```python
counter_list, tag_values, matched_tags = getdata(posts['Title'], tags['TagName'], posts)
```

```python
posts['Recommended_Tags'] = counter_list
posts['Recommended_words'] = tag_values
posts['Exact_Match'] = matched_tags
```
